##Conditional Probability On Different Variables

```python
# Product vs. Gender
product_gender_ct = pd.crosstab(df['Product'], df['Gender'])
print("Contingency Table: Product vs. Gender\n", product_gender_ct)

# Conditional Probability: P(Gender | Product)
product_gender_prob_gender_given_product = product_gender_ct.div(product_gender_ct.sum(axis=1), axis=0)
print("\nConditional Probability (Gender | Product):\n", product_gender_prob_gender_given_product)

# Conditional Probability: P(Product | Gender)
product_gender_prob_product_given_gender = product_gender_ct.div(product_gender_ct.sum(axis=0), axis=1)
print("\nConditional Probability (Product | Gender):\n", product_gender_prob_product_given_gender)

# Marginal Probabilities
product_gender_prob_product = product_gender_ct.sum(axis=1) / product_gender_ct.sum().sum()
print("\nMarginal Probability of Product:\n", product_gender_prob_product)

product_gender_prob_gender = product_gender_ct.sum(axis=0) / product_gender_ct.sum().sum()
print("\nMarginal Probability of Gender:\n", product_gender_prob_gender)
```

Based on the conditional probability analysis of 'Product' and 'Gender', here are the insights and recommendations:

**Insights:**

1.  **Gender Distribution within Products (P(Gender | Product)):**
    *   **KP281:** This product has a relatively balanced gender distribution, with slightly more females (around 51.3%) than males (around 48.8%).
    *   **KP481:** Similar to KP281, this product also has a relatively balanced gender distribution, with a slight tilt towards males (around 51.4%) compared to females (around 48.6%). The difference is minimal for both KP281 and KP481, suggesting these products appeal broadly to both genders.
    *   **KP781:** This product shows a significant skew towards males (around 82.8%) compared to females (around 17.2%). This indicates that the higher-end KP781 treadmill is overwhelmingly purchased by male customers in this dataset.

2.  **Product Distribution within Gender (P(Product | Gender)):**
    *   **Female Customers:** Females are most likely to purchase the KP281 (around 57.3%), followed by the KP481 (around 37.3%), and are least likely to purchase the KP781 (only around 5.3%). This confirms that the lower to mid-range products are much more popular among females.
    *   **Male Customers:** Males are most likely to purchase the KP781 (around 42.7%), followed by the KP281 (around 32.5%), and then the KP481 (around 24.8%). This starkly contrasts with females, showing a strong preference among males for the high-end KP781, while the entry-level KP281 is still a significant choice.

3.  **Marginal Probabilities:**
    *   The marginal probability of purchasing the KP281 is the highest (around 44.4%), followed by KP481 (around 33.3%), and then KP781 (around 22.2%). This aligns with the product pricing (assuming KP281 is the cheapest, KP481 mid-range, and KP781 most expensive), suggesting higher volume sales for the lower-cost models.
    *   The marginal probability of having a female customer is around 41.7%, and a male customer is around 58.3%. The dataset has more male customers overall.

**Key Takeaway:** The KP781 product has a strong gender bias towards male customers, while the KP281 and KP481 are more balanced, with females showing a preference for the lower-end models.
