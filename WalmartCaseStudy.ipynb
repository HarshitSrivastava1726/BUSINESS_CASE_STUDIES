##How does gender affect the amount spent?

```python
mean_purchase_by_marital_status = df.groupby('Marital_Status')['Purchase'].mean()
sem_purchase_by_marital_status = df.groupby('Marital_Status')['Purchase'].sem()

from scipy import stats

for marital_status in mean_purchase_by_marital_status.index:
    mean_purchase = mean_purchase_by_marital_status[marital_status]
    sem_purchase = sem_purchase_by_marital_status[marital_status]
    n = len(df[df['Marital_Status'] == marital_status])
    confidence_interval = stats.t.interval(0.95, n - 1, loc=mean_purchase, scale=sem_purchase)
    print(f"Marital Status: {marital_status}")
    print(f"Mean Purchase: {mean_purchase:.2f}")
    print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})\n")
```

##Confidence Interval Of Purchase amount by Gender

```python

# Calculate mean and SEM for each gender
mean_purchase_by_gender = df.groupby('Gender')['Purchase'].mean()
sem_purchase_by_gender = df.groupby('Gender')['Purchase'].sem()
# Calculate the confidence intervals for each gender
print("Confidence Intervals for Purchase Amount by Gender (Entire Dataset):")
for gender in mean_purchase_by_gender.index:
    mean_purchase = mean_purchase_by_gender[gender]
    sem_purchase = sem_purchase_by_gender[gender]
    n = len(df[df['Gender'] == gender])
    # Using t-distribution for confidence interval calculation
    confidence_interval = stats.t.interval(0.95, n - 1, loc=mean_purchase, scale=sem_purchase)
    interval_width = confidence_interval[1] - confidence_interval[0]
    print(f"Gender: {gender}")
    print(f"Mean Purchase: {mean_purchase:.2f}")
    print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
    print(f"Interval Width: {interval_width:.2f}\n")
```


#### Analysis based on the output:
##### Compare the 'Interval Width' for 'M' and 'F'.
##### The confidence interval is likely narrower for the gender with a larger sample size (more transactions).
##### This is because the Standard Error of the Mean (SEM) is calculated as SEM = Standard Deviation / sqrt(n),
##### where n is the sample size. A larger n results in a smaller SEM, and thus a narrower confidence interval.
##### From previous analysis, we know males have a significantly larger number of transactions, which will lead to a narrower CI for males compared to females.
##### As explained above, the width of the confidence interval is inversely related to the square root of the sample size.
##### A larger sample size (n) leads to a smaller Standard Error of the Mean (SEM), resulting in a narrower confidence interval.
##### Conversely, a smaller sample size leads to a larger SEM and a wider confidence interval.



##Confidence Interval for different Sample Sizes

```python
sample_sizes = [100, 500, 1000, 5000, 10000]
confidence_intervals_by_sample_size = {}

print("Confidence Intervals for Different Sample Sizes:")
for size in sample_sizes:
    # Take a random sample of the specified size
    if size > len(df):
        print(f"Sample size {size} is larger than the dataset size. Skipping.")
        continue
    sample_purchase = df['Purchase'].sample(size, random_state=42) # Use random_state for reproducibility

    # Calculate mean, SEM, and confidence interval for the sample
    sample_mean = sample_purchase.mean()
    sample_sem = stats.sem(sample_purchase)
    # Using t-distribution for confidence interval calculation
    sample_ci = stats.t.interval(0.95, size - 1, loc=sample_mean, scale=sample_sem)
    interval_width = sample_ci[1] - sample_ci[0]

    confidence_intervals_by_sample_size[size] = sample_ci
    print(f"Sample Size: {size}")
    print(f"Sample Mean Purchase: {sample_mean:.2f}")
    print(f"95% Confidence Interval: ({sample_ci[0]:.2f}, {sample_ci[1]:.2f})")
    print(f"Interval Width: {interval_width:.2f}\n")

# Check for overlap between confidence intervals of different sample sizes
sizes_list = list(confidence_intervals_by_sample_size.keys())
print("Checking for Overlap:")
for i in range(len(sizes_list)):
    for j in range(i + 1, len(sizes_list)):
        size1 = sizes_list[i]
        size2 = sizes_list[j]
        ci1 = confidence_intervals_by_sample_size[size1]
        ci2 = confidence_intervals_by_sample_size[size2]

        # Check if intervals overlap
        # Overlap occurs if the upper bound of one is greater than the lower bound of the other AND
        # the lower bound of one is less than the upper bound of the other.
        overlap = (ci1[1] > ci2[0]) and (ci2[1] > ci1[0])

        print(f"Intervals for Sample Sizes {size1} and {size2}: {ci1} and {ci2}")
        print(f"Do they overlap? {'Yes' if overlap else 'No'}")
```

#### Analysis based on the output:
##### Generally, if the confidence intervals are calculated from samples of the *same* population,
##### they are expected to overlap, especially if the sample sizes are not extremely small.
##### Larger sample sizes tend to have narrower intervals, making overlap less likely if the true means are far apart,
##### but in this case, all samples are from the same distribution of 'Purchase'.
##### You will likely see overlap between these confidence intervals

```python


# Let's simulate sampling distributions of means for different sample sizes.

# Define a function to simulate the sampling distribution of the mean
def simulate_sampling_distribution(data, sample_size, num_samples):
    sample_means = [data.sample(sample_size, replace=True).mean() for _ in range(num_samples)]
    return sample_means

num_simulations = 1000 # Number of sample means to calculate for each sample size
sample_sizes_for_clt = [10, 50, 500] # Different sample sizes to demonstrate CLT

plt.figure(figsize=(15, 5))

print("\nSimulating Sampling Distributions of Means:")
for i, size in enumerate(sample_sizes_for_clt):
    if size > len(df):
        print(f"Sample size {size} is larger than the dataset size. Skipping simulation.")
        continue
    simulated_means = simulate_sampling_distribution(df['Purchase'], size, num_simulations)

    plt.subplot(1, len(sample_sizes_for_clt), i + 1)
    sns.histplot(simulated_means, kde=True)
    plt.title(f'Distribution of Sample Means (n={size})')
    plt.xlabel('Sample Mean Purchase')
    plt.ylabel('Frequency')
    print(f"Simulated distribution of means for n={size}. Mean of sample means: {np.mean(simulated_means):.2f}, Std Dev of sample means (SEM): {np.std(simulated_means):.2f}")

plt.tight_layout()
plt.show()

```

##Analysis based on the plots:
#### Observe the shape of the histograms for different sample sizes.
#### For small sample sizes (e.g., n=10), the distribution of sample means might resemble the original population distribution or be somewhat irregular.
#### As the sample size increases (e.g., n=50, n=500), the distribution of sample means will become increasingly bell-shaped (approximating a normal distribution).
#### The spread (standard deviation) of the distribution of sample means will also decrease as the sample size increases (this is the SEM effect).


#### In summary:
#### - i. The CI is narrower for the gender with more transactions (males in this dataset) due to the larger sample size.
#### - ii. Increasing the sample size decreases the width of the confidence interval.
#### - iii. When drawing samples from the same population, confidence intervals of different sample sizes will likely overlap.
#### - iv. As the sample size increases, the distribution of sample means becomes more normally distributed and its spread decreases (Central Limit Theorem).

##Confidence Intervals for Purchase Amount by Age Group

```python
# Calculate mean and SEM for each Age group
mean_purchase_by_age = df.groupby('Age')['Purchase'].mean()
sem_purchase_by_age = df.groupby('Age')['Purchase'].sem()

# Calculate the confidence intervals for each Age group
print("\nConfidence Intervals for Purchase Amount by Age Group (Entire Dataset):")
age_confidence_intervals_full_data = {}
for age_group in mean_purchase_by_age.index:
    mean_purchase = mean_purchase_by_age[age_group]
    sem_purchase = sem_purchase_by_age[age_group]
    n = len(df[df['Age'] == age_group])
    if n > 1: # Ensure sample size is greater than 1 for t-distribution
        # Using t-distribution for confidence interval calculation
        confidence_interval = stats.t.interval(0.95, n - 1, loc=mean_purchase, scale=sem_purchase)
        interval_width = confidence_interval[1] - confidence_interval[0]
        age_confidence_intervals_full_data[age_group] = confidence_interval
        print(f"Age Group: {age_group}")
        print(f"Mean Purchase: {mean_purchase:.2f}")
        print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
        print(f"Interval Width: {interval_width:.2f}\n")
    else:
        print(f"Age Group: {age_group}")
        print(f"Sample size is 1 or less. Cannot compute confidence interval.\n")

```

##Overlap of Confidence Intervals for different sample sizes for Age Groups

```python
# iii. Do the confidence intervals for different sample sizes overlap?
print("\nChecking for Overlap of Confidence Intervals for different sample sizes (for Age groups):")
# Let's use a single age group to demonstrate the effect of sample size on CI overlap.
# We'll pick a representative age group, e.g., '26-35'.
age_group_to_analyze = '26-35'
if age_group_to_analyze in df['Age'].unique():
    df_age_subset = df[df['Age'] == age_group_to_analyze]['Purchase']

    sample_sizes_age = [100, 500, 1000] # Different sample sizes for demonstration
    confidence_intervals_age_sample_size = {}

    print(f"Confidence Intervals for Age Group '{age_group_to_analyze}' with Different Sample Sizes:")
    for size in sample_sizes_age:
        # Take a random sample of the specified size from the age group subset
        if size > len(df_age_subset):
            print(f"Sample size {size} is larger than the subset size for Age Group '{age_group_to_analyze}'. Skipping.")
            continue
        sample_purchase_age = df_age_subset.sample(size, random_state=42) # Use random_state for reproducibility

        # Calculate mean, SEM, and confidence interval for the sample
        sample_mean_age = sample_purchase_age.mean()
        sample_sem_age = stats.sem(sample_purchase_age)
        # Using t-distribution for confidence interval calculation
        sample_ci_age = stats.t.interval(0.95, size - 1, loc=sample_mean_age, scale=sample_sem_age)
        interval_width_age = sample_ci_age[1] - sample_ci_age[0]

        confidence_intervals_age_sample_size[size] = sample_ci_age
        print(f"Sample Size (within Age Group '{age_group_to_analyze}'): {size}")
        print(f"Sample Mean Purchase: {sample_mean_age:.2f}")
        print(f"95% Confidence Interval: ({sample_ci_age[0]:.2f}, {sample_ci_age[1]:.2f})")
        print(f"Interval Width: {interval_width_age:.2f}\n")

    # Check for overlap between confidence intervals of different sample sizes for this age group
    sizes_list_age = list(confidence_intervals_age_sample_size.keys())
    print(f"Checking for Overlap for Age Group '{age_group_to_analyze}':")
    for i in range(len(sizes_list_age)):
        for j in range(i + 1, len(sizes_list_age)):
            size1 = sizes_list_age[i]
            size2 = sizes_list_age[j]
            ci1 = confidence_intervals_age_sample_size[size1]
            ci2 = confidence_intervals_age_sample_size[size2]

            # Check if intervals overlap
            overlap = (ci1[1] > ci2[0]) and (ci2[1] > ci1[0])

            print(f"Intervals for Sample Sizes {size1} and {size2}: {ci1} and {ci2}")
            print(f"Do they overlap? {'Yes' if overlap else 'No'}")

else:
    print(f"Age group '{age_group_to_analyze}' not found in the dataset.")

```
